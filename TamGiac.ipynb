```python
# Import
import math
import turtle

# TAM GIÁC

# Input tọa độ
xa = xA = int(input("xA = "))
ya = yA = int(input("yA = "))
xb = xB = int(input("xB = "))
yb = yB = int(input("yB = "))
xc = xC = int(input("xC = "))
yc = yC = int(input("yC = "))

# Xác định có phải tam giác không
tamgiac = (xa != xb or xb != xc or xc != xa) and (ya != yb or yb != yc or yc != ya)
print("Tam giác là tác giam: ", tamgiac)

# Tìm độ dài các cạnh
AB = BA = math.sqrt((xb-xa)**2 + (yb-ya)**2)
AC = CA = math.sqrt((xc-xa)**2 + (yc-ya)**2)
BC = CB = math.sqrt((xc-xb)**2 + (yc-yb)**2)

# Tìm góc các cạnh
cosA = ((AB**2 + AC**2 - BC**2)/(2 * AB * AC))
cosB = ((BA**2 + BC**2 - AC**2)/(2 * BA * BC))
cosC = ((CB**2 + CA**2 - BA**2)/(2 * CB * CA))

print(cosC)

gA = math.degrees(math.acos(cosA))
gB = math.degrees(math.acos(cosB))
gC = math.degrees(math.acos(cosC))

# Tìm độ dài đường cao đường cao
A_caoA = caoA_A = math.sin(gC)*AC
B_caoB = caoB_B = math.sin(gC)*BC
C_caoC = caoC_C = math.sin(gB)*BC

# Print độ dài đường cao
print("Độ dài đường cao từ đỉnh A là: " + str(A_caoA))
print("Độ dài đường cao từ đỉnh B là: " + str(B_caoB))
print("Độ dài đường cao từ đỉnh C là: " + str(C_caoC))

# Tìm trung tuyến
xm = xM = x_midpoint_AB = (xa + xb)/2
ym = yM = y_midpoint_AB = (ya + yb)/2
xn = xN = x_midpoint_AC = (xa + xc)/2
yn = yN = y_midpoint_AC = (ya + yc)/2
xo = xO = x_midpoint_BC = (xb + xc)/2
yo = yO = y_midpoint_BC = (yb + yc)/2

# Độ dài trung tuyến
CM = MC = math.sqrt((xm-xc)**2 + (ym-yc)**2)
BN = NB = math.sqrt((xn-xb)**2 + (yn-yb)**2)
AO = OA = math.sqrt((xo-xa)**2 + (yo-ya)**2)

# Print độ dài trung tuyến
print("Độ dài đường trung tuyến từ A (AO) là: " + str(OA))
print("Độ dài đường trung tuyến từ B (BN) là: " + str(BN))
print("Độ dài đường trung tuyến từ C (CM) là: " + str(CM))

# Tọa độ trọng tâm
xi = xI = x_center_ABC = (xa + xb + xc)/3
yi = yI = y_center_ABC = (ya + yb + yc)/3

# Print tọa độ trọng tâm
print("Tọa độ trọng tâm của tam giác ABC là: (" + str(xi) + "; " + str(yi) + ")")


# VISUALIZE

# Setting cho rùa
t=turtle.Turtle()
t.shape("turtle")
t.pensize(1)

# Vẽ tam giác ABC
t.penup()
t.goto(xA, yA)
t.pendown()
t.goto(xB, yB)
t.goto(xC, yC)
t.goto(xA, yA)
t.penup()
t.goto(xA, yA)
t.pendown()
t.goto(xO, yO)
t.penup()
t.goto(xB, yB)
t.pendown()
t.goto(xN, yN)
t.penup()
t.goto(xC, yC)
t.pendown()
t.goto(xM, yM)
t.penup()
t.goto(xI, yI)
t.pendown()

# Vẽ xong
turtle.done
```
